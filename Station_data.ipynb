```
#Load your json file into an object
with urllib.request.urlopen("https://offenedaten-wuppertal.de/sites/default/files/Schwebebahnhoefe_EPSG4326_JSON.json") as url:
    train_data = json.loads(url.read().decode())
```

```
#Cleaning: We want to keep only data after 'features'    
station_data = train_data['features']
```

```
#Create your dataframe
results = pd.DataFrame()
```

```
#Loop to store datas into the dataframe
for data in station_data:

    station_name = data['properties']['NAME']
    temp_df = pd.DataFrame(data['geometry']['coordinates'])
    temp_df = temp_df.T


    temp_df['StationName'] = station_name

    results = results.append(temp_df).reset_index(drop=True)
    
results.head()
```

```
#We want to reorganize the colums to have Station Name first and keep the order after (latitude then longitude)
df_station = results[['StationName',1,0]]
df_station.head()
```

```
#Rename some columns
df_station.rename(columns={1:'StationLatitude',
                       0:'StationLongitude'       
                      }, 
                 inplace=True)
df_station
```
